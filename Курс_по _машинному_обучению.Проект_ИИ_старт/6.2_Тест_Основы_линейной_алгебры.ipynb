## Найдите значение x+y, если даны точки A(x; -3), B(2; -7) и вектор AB(5; y).


##### Чтобы найти координаты вектора на плоскости, если он задан координатами своих начала и конца , необходимо от координат конца отнять соответствующие координаты начала.

2 - x = 5        

x = 2 - 5 = -3         

                     
-7 - (-3) = y     

y = - 4

x + y = -3  + (-4) = -7

##### По определению вектора: если даны две точки A(x1, y1) и B(x2, y2), то будет вектор AB(x2-x1, y2-y1). Порядок точек в векторе ВАЖЕН!

Для A(x;−3), B(2; -7) и вектора AB(5; y) имеем уравнения: 2 - x = 5 и -7 + 3 = y.

Отсюда находим: x = -3 и y = -4.

Ответ: x + y = -3 + -4 = -7
