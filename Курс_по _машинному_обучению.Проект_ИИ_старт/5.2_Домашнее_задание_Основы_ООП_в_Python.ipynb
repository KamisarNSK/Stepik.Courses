## Задача простая и немножко на внимательность. Дольше думаешь над расшифровкой массив весов нейрона и т.д.

Реализуйте класс "Нейрон", у которого будет несколько методов: 

    -  __init__. Принимает на вход список/кортеж с числами w , [w1, w2, w3....и т.д.] , и  некую функцию f.
    - forward. Принимает на вход список/кортеж с числами x, [x1, x2, x3....и т.д.], и возвращает результат выполнения функции f аргументом в которую передаётся сумма перемноженных между собой элементов массивов w и x. например return f(sum([x1 * w1, w2 * x2, w3 * x3....и т.д.]))
    - backlog. Возвращает список/кортеж x, который подавался на вход функции forward при её последнем вызове. Если ранее функция forward не вызывалось, возвращает None.

```python
class Neuron:
    prev_x = list()
    def __init__(self, w, f = lambda x: x):
        self.weight = w
        self.func = f
        prev_x = None
    def forward(self, x):
        self.prev_x = x
        result = sum(map(lambda i1, i2: i1 * i2, self.weight, x))
        return self.func(result)
    def backlog(self):
        return self.prev_x
```

```python
# чужое
class Neuron:

    def __init__(self, w, f=lambda x: x):
        self.weights = w
        self.function = f
        self.__last_input = None

    def forward(self, x):
        if not isinstance(x, (list, tuple, frozenset)):
            raise TypeError('на входе должен быть массив')            
        if len(x) > len(self.weights):
            raise IndexError('количество входов превышает количество весов')
            
        self.__last_input = list(x)
        extra_len = len(self.weights) - len(x)
        if extra_len:
            x += [0] * extra_len
        return self.function(sum(w * i for w, i in zip(self.weights, x)))

    def backlog(self):
        return self.__last_input
```
