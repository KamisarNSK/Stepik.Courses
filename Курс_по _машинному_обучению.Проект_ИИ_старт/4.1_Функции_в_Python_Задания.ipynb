## Почти двойной факториал

    Реализуйте функцию almost_double_factorial(n), вычисляющую произведение всех нечётных натуральных чисел, не превосходящих nn.
    В качестве аргумента ей передаётся натуральное (ноль -- натуральное) число n \leqslant 100n⩽100.
    Возвращаемое значение - вычисленное произведение.

Комментарий. В случае, если n = 0, требуется вернуть 1.

В этом задании функция print вам не понадобится. Результаты выполнения функций нужно возвращать, а не печатать!

```python
def almost_double_factorial(n):
    count=1
    for i in range(1,n+1):
        if i%2:
            count *= i        
    return count
```

```python
almost_double_factorial(0)
```

```python
# чужое
def almost_double_factorial(n):
  p = 1
  for i in range(1, n+1, 2):
    p = p*i
  return p
```

```python
# чужое
def almost_double_factorial(n):
    return __import__('functools').reduce(lambda x, y: x * y, filter(lambda x: x % 2, range(1, n + 1)), 1)
```

```python
# чужое
from functools import reduce
from operator import mul
def almost_double_factorial(n):
    if n == 0:
        return 1
    return reduce(mul, range(1, n + 1, 2))
```

## Реализуйте функцию calls_num(), которая будет возвращать число предыдущих вызовов данной функции.

```python
count_call = -1
def calls_num():
    global count_call
    count_call +=1
    return count_call
```

## Реализуйте функцию cmp_prev(val), которая будет возвращать True , если значение val больше суммы двух предыдущих значений val при вызове той же функции, и False иначе.

```python
sum_fun1 = 0
sum_fun2 = 0
def cmp_prev(val):
    flag = True
    global sum_fun1, sum_fun2
    if val <= sum_fun2+sum_fun1:
        flag = False
    sum_fun1, sum_fun2 = sum_fun2, val
    return flag
```

```python
cmp_prev(4)
```

```python
# чужое
a  = [0,0]
def cmp_prev(val):
    a.append(val)
    return(a[-2] + a[-3] < val) 
```

```python
# чужое
l = [0, 0]
def cmp_prev(val):
    global l
    r = val > sum(l)
    del l[0]
    l.append(val)
    #print(val, l, sum(l), val > sum(l))
    return r
```

```python
# чужое
numb  = []

def cmp_prev(val):  
    numb.append(val)
    return(val > sum(numb[::-1][1:3]))
```

```python
# чужое
# Будто бы не помешает чистить глобальный список, чтоб в памяти не висел объект весом со слона.
prev = [0, 0, 0]

def cmp_prev(val):
    global prev
    if len(prev) >= 3:
        prev.pop(0)
    prev.append(val)

    return val > sum(prev[0:2])
```

```python
# чужое
def Set(func):
    global new_func
    new_func = func

def Compute(string):
    return new_func(string)
```
