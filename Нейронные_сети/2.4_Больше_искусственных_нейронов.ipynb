```python
from sympy import diff
import math
import numpy as np
from sympy import Symbol, Derivative


def sigmoid_derivative(a):
    x= Symbol('x')
    function=  1 / (1 + math.e ** -x) 
    deriv= Derivative(function, x)
    return deriv.doit().subs({x:a})

def tanh_derivative(a):
    x= Symbol('x')
    function=  (math.e ** x - math.e ** -x)/(math.e ** x + math.e ** -x)
    deriv= Derivative(function, x)
    return deriv.doit().subs({x:a})

def tanh(x):
    a = np.tanh(x)
    return a

def sigmoid(x):
     return 1 / (1 + math.e ** -x) 

print(sigmoid_derivative(0))
print(tanh_derivative(0))
print(tanh(-1))
print(tanh(1))
print(1 - tanh(1))
print(sigmoid(-1))
print(sigmoid(1))
print(1 - sigmoid(1))
```
